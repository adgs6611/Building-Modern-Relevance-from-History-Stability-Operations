```python
import networkx as nx  # For creating and managing the network graph
import plotly.graph_objects as go  # For creating interactive visualizations

# Step 1: Define the relationships between entities as directed edges
# Each tuple (source, target) represents a directed relationship
relationships = [
    ("SCAP", "Japanese Government"),  # SCAP oversees the Japanese Government
    ("Japanese Government", "Ministry of Education"),  # Japanese Government supervises the Ministry of Education
    ("Japanese Government", "Ministry of Health"),  # Japanese Government supervises the Ministry of Health
    ("Japanese Government", "Ministry of Agriculture"),  # Japanese Government supervises the Ministry of Agriculture
    ("Ministry of Education", "Bureau of Curriculum"),  # Ministry of Education oversees Bureau of Curriculum
    ("Ministry of Health", "Bureau of Public Welfare"),  # Ministry of Health oversees Bureau of Public Welfare
    ("Ministry of Agriculture", "Bureau of Food Production"),  # Ministry of Agriculture oversees Bureau of Food Production
    ("SCAP", "Economic and Scientific Section"),  # SCAP manages Economic and Scientific Section
    ("SCAP", "Civil Information and Education Section"),  # SCAP manages Civil Information and Education Section
    ("Economic and Scientific Section", "Ministry of Finance"),  # Economic section connects to Ministry of Finance
    ("Civil Information and Education Section", "Ministry of Education"),  # Civil section connects to Ministry of Education
]

# Step 2: Define a description for each organization
# These descriptions will be displayed as hover text in the graph
descriptions = {
    "SCAP": "Supreme Commander for the Allied Powers, oversaw post-war Japan.",
    "Japanese Government": "Central governing body of Japan during the occupation.",
    "Ministry of Education": "Supervised educational reforms and curriculum development.",
    "Ministry of Health": "Managed public health and welfare systems.",
    "Ministry of Agriculture": "Focused on agricultural recovery and food production.",
    "Economic and Scientific Section": "Handled economic stabilization and industrial recovery.",
    "Civil Information and Education Section": "Implemented civil education and media reforms.",
    "Bureau of Curriculum": "Developed and revised school curricula.",
    "Bureau of Public Welfare": "Oversaw welfare programs for public health.",
    "Bureau of Food Production": "Ensured food supply and agricultural development.",
    "Ministry of Finance": "Managed financial policies and economic stabilization.",
}

# Step 3: Create a directed graph using NetworkX
G = nx.DiGraph()  # Initialize a directed graph
G.add_edges_from(relationships)  # Add all relationships as edges in the graph

# Step 4: Define hierarchical positions for nodes using Graphviz
# 'dot' layout generates a top-down hierarchical arrangement
pos = nx.nx_agraph.graphviz_layout(G, prog="dot")

# Step 5: Extract edge positions for visualization
edge_x = []  # x-coordinates of edges
edge_y = []  # y-coordinates of edges
for edge in G.edges():  # Iterate through all edges in the graph
    x0, y0 = pos[edge[0]]  # Start position of the edge
    x1, y1 = pos[edge[1]]  # End position of the edge
    edge_x.append(x0)  # Add x-coordinate of the start
    edge_x.append(x1)  # Add x-coordinate of the end
    edge_x.append(None)  # Add a None to break between edges
    edge_y.append(y0)  # Add y-coordinate of the start
    edge_y.append(y1)  # Add y-coordinate of the end
    edge_y.append(None)  # Add a None to break between edges

# Step 6: Create edge traces for Plotly
edge_trace = go.Scatter(
    x=edge_x,  # x-coordinates of edges
    y=edge_y,  # y-coordinates of edges
    line=dict(width=1, color="gray"),  # Style of the edges
    hoverinfo="none",  # No hover text for edges
    mode="lines"  # Draw as lines
)

# Step 7: Extract node positions, labels, and attributes
node_x = []  # x-coordinates of nodes
node_y = []  # y-coordinates of nodes
node_text = []  # Hover text for nodes
node_color = []  # Color of nodes
node_size = []  # Size of nodes

for node in G.nodes():  # Iterate through all nodes in the graph
    x, y = pos[node]  # Get the position of the node
    node_x.append(x)  # Add x-coordinate of the node
    node_y.append(y)  # Add y-coordinate of the node
    node_text.append(f"{node}: {descriptions[node]}")  # Add detailed description for hover text
    node_color.append(  # Assign color based on the type of node
        "red" if node == "SCAP" else
        "orange" if node == "Japanese Government" else
        "blue" if "Ministry" in node else
        "purple"
    )
    node_size.append(  # Assign size based on the importance of the node
        20 if node == "SCAP" else
        15 if "Government" in node else
        10
    )

# Step 8: Create node traces for Plotly
node_trace = go.Scatter(
    x=node_x,  # x-coordinates of nodes
    y=node_y,  # y-coordinates of nodes
    mode="markers+text",  # Show markers and labels
    text=[node for node in G.nodes()],  # Node names as labels
    hovertext=node_text,  # Detailed descriptions for hover text
    hoverinfo="text",  # Show only hover text on hover
    marker=dict(  # Style for the markers
        color=node_color,  # Node colors
        size=node_size,  # Node sizes
        line=dict(width=2, color="black")  # Black border for nodes
    ),
    textposition="bottom center"  # Position labels below nodes
)

# Step 9: Create the Plotly figure
fig = go.Figure(data=[edge_trace, node_trace],  # Add edge and node traces
                layout=go.Layout(
                    title="Hierarchical Network of Relationships: SCAP and Japanese Government",  # Title of the graph
                    titlefont_size=16,  # Font size for the title
                    showlegend=False,  # No legend is needed
                    hovermode="closest",  # Highlight the closest point on hover
                    margin=dict(b=0, l=0, r=0, t=40),  # Set margins for the plot
                    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),  # Hide x-axis
                    yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)   # Hide y-axis
                ))

# Step 10: Display the figure
fig.show()  # Render the interactive network graph
```

```python
# Import necessary libraries
from sklearn.feature_extraction.text import TfidfVectorizer  # For extracting important keywords
from sklearn.cluster import KMeans  # For clustering keywords into categories
import pandas as pd  # For managing data in tabular format
import nltk  # For natural language processing
from nltk.corpus import stopwords  # For removing common words
import re  # For regular expressions
import string  # For handling punctuation
from docx import Document  # To read .docx files

# Ensure NLTK resources are downloaded
nltk.download('stopwords')  # Download the stopwords corpus

# Step 1: Define the file paths for the documents
# Specify the directory where the documents are located
directory = "/Users/g1/Documents/python_files"
file1 = os.path.join(directory, "Non_military_activities No.1.docx")
file2 = os.path.join(directory, "Non-Military Activities No.2.docx")
file3 = os.path.join(directory, "Non-Military Activities No.3.docx")

# List of file paths
file_paths = [file1, file2, file3]

# Step 2: Function to read text from .docx files
def read_docx(file_path):
    """
    Reads a .docx file and extracts all text paragraphs.
    """
    doc = Document(file_path)  # Load the Word document
    return "\n".join([para.text for para in doc.paragraphs])  # Combine all paragraphs into a single string

# Read all documents into a list
documents = [read_docx(file_path) for file_path in file_paths]

# Step 3: Preprocess the text
def preprocess_text(text):
    """
    Cleans and preprocesses the text:
    - Removes punctuation and digits
    - Converts to lowercase
    - Removes stopwords
    - Tokenizes the text
    """
    text = re.sub(r"[{}\d]".format(string.punctuation), " ", text)  # Remove punctuation and digits
    text = text.lower()  # Convert text to lowercase
    stop_words = set(stopwords.words('english'))  # Define stopwords
    tokens = nltk.word_tokenize(text)  # Tokenize text into words
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]  # Remove stopwords and short words
    return " ".join(tokens)  # Return cleaned text as a single string

# Preprocess each document
cleaned_documents = [preprocess_text(doc) for doc in documents]

# Step 4: Extract keywords using TF-IDF
tfidf = TfidfVectorizer(max_features=1000)  # Limit to the top 1000 keywords
tfidf_matrix = tfidf.fit_transform(cleaned_documents)  # Fit the TF-IDF model to the cleaned documents
terms = tfidf.get_feature_names_out()  # Extract the keywords

# Step 5: Categorize keywords using K-Means clustering
num_clusters = 4  # Define the number of clusters (Economic, Social, Political, Military)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)  # Initialize K-Means clustering
kmeans.fit(tfidf_matrix.T)  # Fit the clustering model on the features (transpose matrix to cluster terms)

# Create a DataFrame to map terms to clusters
keywords_df = pd.DataFrame({
    "Keyword": terms,  # The extracted keywords
    "Category": kmeans.labels_  # The cluster labels assigned by K-Means
})

# Step 6: Map cluster numbers to meaningful category names
category_labels = {0: "Economic", 1: "Social", 2: "Political", 3: "Military"}
keywords_df["Category"] = keywords_df["Category"].map(category_labels)  # Replace numeric labels with category names

# Step 7: Summarize top concerns for each category
top_keywords = keywords_df.groupby("Category")["Keyword"].apply(list)  # Group keywords by category

# Step 8: Display the results using pandas
from IPython.display import display  # For displaying DataFrames in Jupyter
print("Top Keywords Categorized:")
display(keywords_df)  # Display the DataFrame interactively

# Print the summary of top keywords by category
print("Summary of Top Concerns by Category:")
for category, keywords in top_keywords.items():
    print(f"\n{category}:\n{', '.join(keywords)}")
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re
import string
from docx import Document

# Ensure NLTK resources are downloaded
nltk.download('stopwords')

# Step 1: Define file paths for the documents
directory = "/Users/g1/Documents/python_files"
file1 = os.path.join(directory, "Non_military_activities No.1.docx")
file2 = os.path.join(directory, "Non-Military Activities No.2.docx")
file3 = os.path.join(directory, "Non-Military Activities No.3.docx")

file_paths = [file1, file2, file3]

# Function to read text from Word documents
def read_docx(file_path):
    """
    Reads a .docx file and extracts all text paragraphs.
    """
    doc = Document(file_path)
    return "\n".join([para.text for para in doc.paragraphs])

# Read all documents into a list
documents = [read_docx(file_path) for file_path in file_paths]

# Step 2: Preprocess the text
def preprocess_text(text):
    """
    Cleans and preprocesses the text:
    - Removes punctuation and digits
    - Converts to lowercase
    - Removes stopwords
    - Tokenizes the text
    """
    text = re.sub(r"[{}\d]".format(string.punctuation), " ", text)  # Remove punctuation and digits
    text = text.lower()  # Convert text to lowercase
    stop_words = set(stopwords.words('english'))  # Define stopwords
    tokens = nltk.word_tokenize(text)  # Tokenize text into words
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]  # Remove stopwords and short words
    return tokens  # Return tokens as a list

# Preprocess all documents
all_tokens = []
for doc in documents:
    all_tokens.extend(preprocess_text(doc))  # Combine tokens from all documents

# Step 3: Perform frequency analysis
word_counts = Counter(all_tokens)  # Count occurrences of each word

# Step 4: Categorize keywords into themes
# Predefined keywords for each category
categories = {
    "Economic": ["industry", "finance", "trade", "inflation", "recovery", "resources", "money", "jobs"],
    "Social": ["education", "health", "welfare", "housing", "food", "population", "care", "refugees"],
    "Political": ["government", "policy", "reform", "democracy", "law", "control", "administration", "diplomacy"],
    "Military": ["troops", "army", "security", "war", "repatriation", "strategy", "peacekeeping", "occupation"],
}

# Map words to their categories
category_counts = {key: 0 for key in categories.keys()}  # Initialize category counts
for word, count in word_counts.items():
    for category, keywords in categories.items():
        if word in keywords:  # Check if the word belongs to a category
            category_counts[category] += count  # Add the word count to the category

# Step 5: Convert category counts to a DataFrame for visualization
category_df = pd.DataFrame(list(category_counts.items()), columns=["Category", "Frequency"])

# Step 6: Visualize the results
# Bar chart to show frequency of pressing problems
plt.figure(figsize=(10, 6))
plt.bar(category_df["Category"], category_df["Frequency"], color=["blue", "green", "orange", "red"])
plt.title("Most Pressing Problems Faced by Allied Forces", fontsize=16)
plt.xlabel("Category", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the category counts as a DataFrame for review
from IPython.display import display
print("Frequency of Pressing Problems by Category:")
display(category_df)
```

```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re
import string
from docx import Document

# Ensure NLTK resources are downloaded
nltk.download('stopwords')  # Download stopwords corpus

# Step 1: Define file paths for the documents
directory = "/Users/g1/Documents/python_files"
file1 = os.path.join(directory, "Non_military_activities No.1.docx")
file2 = os.path.join(directory, "Non-Military Activities No.2.docx")
file3 = os.path.join(directory, "Non-Military Activities No.3.docx")

file_paths = [file1, file2, file3]

# Function to read text from Word documents
def read_docx(file_path):
    """
    Reads a .docx file and extracts all text paragraphs.
    """
    doc = Document(file_path)
    return "\n".join([para.text for para in doc.paragraphs])

# Read all documents into a list
documents = [read_docx(file_path) for file_path in file_paths]

# Step 2: Preprocess the text
def preprocess_text(text):
    """
    Cleans and preprocesses the text:
    - Removes punctuation and digits
    - Converts to lowercase
    - Removes stopwords
    - Tokenizes the text
    """
    text = re.sub(r"[{}\d]".format(string.punctuation), " ", text)  # Remove punctuation and digits
    text = text.lower()  # Convert text to lowercase
    stop_words = set(stopwords.words('english'))  # Define stopwords
    tokens = nltk.word_tokenize(text)  # Tokenize text into words
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]  # Remove stopwords and short words
    return tokens  # Return tokens as a list

# Preprocess all documents
all_tokens = []
for doc in documents:
    all_tokens.extend(preprocess_text(doc))  # Combine tokens from all documents

# Step 3: Perform frequency analysis
word_counts = Counter(all_tokens)  # Count occurrences of each word

# Step 4: Categorize keywords into themes
# Predefined keywords for each category
categories = {
    "Economic": ["industry", "finance", "trade", "inflation", "recovery", "resources", "money", "jobs"],
    "Social": ["education", "health", "welfare", "housing", "food", "population", "care", "refugees"],
    "Political": ["government", "policy", "reform", "democracy", "law", "control", "administration", "diplomacy"],
    "Military": ["troops", "army", "security", "war", "repatriation", "strategy", "peacekeeping", "occupation"],
}

# Step 5: Map words to their categories
category_words = {key: [] for key in categories.keys()}  # Initialize a dictionary to store words for each category
for word, count in word_counts.items():
    for category, keywords in categories.items():
        if word in keywords:  # Check if the word belongs to a category
            category_words[category].append((word, count))  # Add the word and its frequency to the category

# Step 6: Identify top problems in each category
top_problems = {}
for category, words in category_words.items():
    top_problems[category] = sorted(words, key=lambda x: x[1], reverse=True)[:5]  # Get the top 5 words for each category

# Step 7: Visualize the top problems for each category
for category, problems in top_problems.items():
    # Convert problems to a DataFrame for easier handling
    df = pd.DataFrame(problems, columns=["Problem", "Frequency"])

    # Create a bar chart
    plt.figure(figsize=(8, 5))
    plt.bar(df["Problem"], df["Frequency"], color="skyblue")
    plt.title(f"Top Problems in {category} Category", fontsize=16)
    plt.xlabel("Problems", fontsize=14)
    plt.ylabel("Frequency", fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Step 8: Display the results as a DataFrame for review
for category, problems in top_problems.items():
    print(f"\nTop Problems in {category} Category:")
    df = pd.DataFrame(problems, columns=["Problem", "Frequency"])
    display(df)
```

Below is an example of actionable lesson learned from the Allied Forces reporting at the time. This tool would be useful to provide additional contexts to the problems presented at the time and to evaluate them for modern application.

```python
from docx import Document
import re
import pandas as pd

# Step 1: Define file paths for the documents
directory = "/Users/g1/Documents/python_files"
file1 = f"{directory}/Non_military_activities No.1.docx"
file2 = f"{directory}/Non-Military Activities No.2.docx"
file3 = f"{directory}/Non-Military Activities No.3.docx"
file_paths = [file1, file2, file3]

# Step 2: Function to read text from Word documents
def read_docx(file_path):
    """
    Reads a .docx file and extracts all text paragraphs.
    """
    doc = Document(file_path)
    return "\n".join([para.text.strip() for para in doc.paragraphs if para.text.strip()])

# Combine all document texts into one
documents = [read_docx(file) for file in file_paths]
all_text = " ".join(documents)

# Step 3: Extract lessons based on keywords
def extract_lessons(text):
    """
    Extracts actionable lessons based on keyword patterns.
    """
    lessons = []
    sentences = re.split(r"[.!?]", text)  # Split text into sentences
    keywords = [
        "success", "failure", "lesson", "reform", "challenge", "opportunity", "impact", "strategy"
    ]  # Keywords to identify lessons

    for sentence in sentences:
        if any(keyword in sentence.lower() for keyword in keywords):
            lessons.append(sentence.strip())

    return lessons

# Extract lessons from the documents
lessons = extract_lessons(all_text)

# Step 4: Organize lessons into categories
def categorize_lessons(lessons):
    """
    Categorizes lessons into key themes.
    """
    categories = {
        "Governance": ["policy", "law", "government", "administration"],
        "Food Security": ["food", "supply", "distribution", "hunger", "ration"],
        "Public Health": ["health", "disease", "sanitation", "vaccination", "hygiene"],
        "Economic Recovery": ["economy", "trade", "industry", "finance", "jobs"]
    }
    categorized_lessons = {category: [] for category in categories}

    for lesson in lessons:
        for category, keywords in categories.items():
            if any(keyword in lesson.lower() for keyword in keywords):
                categorized_lessons[category].append(lesson)
                break  # Stop after assigning a category

    return categorized_lessons

# Categorize the lessons
categorized_lessons = categorize_lessons(lessons)

# Step 5: Convert to DataFrame
lesson_data = []
for category, lesson_list in categorized_lessons.items():
    for lesson in lesson_list:
        lesson_data.append({"Category": category, "Lesson": lesson})

df_lessons = pd.DataFrame(lesson_data)

# Step 6: Export the DataFrame to a CSV file
output_file = "lessons_learned.csv"
df_lessons.to_csv(output_file, index=False)
print(f"Lessons have been saved to '{output_file}'.")

# Step 7: Print summary of lessons
print("\nActionable Lessons Summary:")
for category, lesson_list in categorized_lessons.items():
    print(f"\nCategory: {category}")
    for lesson in lesson_list[:5]:  # Limit to 5 lessons per category for brevity
        print(f"- {lesson}")
```
